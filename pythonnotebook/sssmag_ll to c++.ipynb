```python
fftocppif={"IF":"if","THEN":"{","abs":"","lmaxa":"_lmax_alpha","lmaxb":"_lmax_gamma",".and.":" && ",".ge.":">"}
```

```python
def overlapline(line):

#test="S(10,10) = pma(3)*S( 4,10) + fak * (S( 1,10) + 2d0*S( 4, 4) )"


    def Sresult(result):
        pattern="\d+"
        entries=re.findall(pattern,result)
        ol="S[{}][{}][{}]".format(int(entries[0])-1,int(entries[1])-1,int(entries[2])-1)   
        return ol
    
    def pma(result):
        pattern="\d+"
        entries=re.findall(pattern,result)
        pma="gma{}".format(int(entries[0])-1)    
        return pma
    
    def pmb(result):
        pattern="\d+"
        entries=re.findall(pattern,result)
        pmb="gmb{}".format(int(entries[0])-1)    
        return pmb
    
    def pmc(result):
        pattern="\d+"
        entries=re.findall(pattern,result)
        pmb="gmc{}".format(int(entries[0])-1)    
        return pmb
    
    def d(result):
        return result.replace("d","e")
    
    fftocppS={"S\(\s*\d+\s*,\s*\d+\s*,\s*\d+\s*\)":Sresult,"gma\(\d+\)":pma,"gmc\(\d+\)":pmc,"gmb\(\d+\)":pmb,"\d+\.*d\d+":d}
    
    new=line
    for pattern in fftocppS:
        found=re.findall(pattern,new)
        if found:
            #print found
            for f in found:
                new=new.replace(f,fftocppS[pattern](f))
    return new
    
```

```python

def ifline(line):
    result=line
    for string in fftocppif:
        if string in line:
            result=result.replace(string,fftocppif[string])
    return result
```

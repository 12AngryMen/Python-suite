```python
def readSqlall(sqlname):
        sqlstatement = "SELECT Jeff2s FROM pairs"
        con = sqlite3.connect(sqlname)
        with con:
                cur = con.cursor()
                cur.execute(sqlstatement)
                rows = cur.fetchall()

        sqlall=(np.array(rows)[:,0])
        
        return sqlall
    
```

```python
pairs=readSqlall("system_individualmps.sql")
```
